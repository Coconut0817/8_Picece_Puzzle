```python
import random
import copy
from sortedcontainers import SortedList

class Queue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        if self.items ==[]:
            return print('this is empty')
        else :
            return print('裡面還有值')
    def enqueue(self, item): # 從list index = 0 處開始加入item
        self.items.insert(0,item)

    def dequeue(self):       # 從list index = -1 處del item
        return self.items.pop()

    def size(self):
        return len(self.items)
def swap(status, t1, t2):
    c = status[t1]
    status[t1] = status[t2] 
    status[t2] = c
    return Node
    
class Node():
    status = bytearray() # 當前狀態
    father = None # 紀錄上個狀態，如果father = None 該點為根節點
    step = None # 紀錄上個移動，如果Step = None 該點為根節點
    def __init__(self, status, father, step):
        self.status = status
        self.father = father
        self.step = step
    def __len__(self):
        return len(self.status)
    def ToSequence(self):
        result = 0;
        for i in range(0, len(self)):
          result = result * 10 + self.status[i];
        return result;

def GetNext(now):# 輸入當前版面並回傳0所有移動後的狀態
    index = now.status.index(0)
    col = int(index % 3)
    row = int(index / 3)
    nextPush = list()
    next = Node
    if (row != 0): # Top
        next = copy.deepcopy(now.status)   
        swap(next, index, index - 3)#跟上面交換        
        nextPush.append(Node(next, now, 'T'))#加入這個新狀態
    if (col != 2): # Right
        next = copy.deepcopy(now.status)        
        swap(next, index, index + 1);# 跟右邊交換 
        nextPush.append(Node(next, now, 'R'))
    if (row != 2): # Bottom
        next = copy.deepcopy(now.status)          
        swap(next, index, index + 3)# 跟下面交換
        nextPush.append(Node(next, now, 'B'))
    if (col != 0): # Left
        next = copy.deepcopy(now.status)          
        swap(next, index, index - 1)# 跟左邊交換
        nextPush.append(Node(next, now, 'L'))            
    return nextPush

def Solve(source, goal):
    queue = Queue();#使用狀態序列來儲存已走過的路徑，防止往回走
    book = dict();
    end = Node(goal, None, None)# 終點
    start = Node(source, None, None)# 起點
    queue.enqueue(start)# 推入起點
    book.setdefault(start.ToSequence(), []).append(True);
    endStatus = end.ToSequence();
    while (queue.size() > 0):# 取得當前搜索狀態，並移出佇列
        now = Node
        now = queue.dequeue()
        #print(now.status,now.father)
        if (now.ToSequence() == endStatus):
          return PathTrace(now) # 如果抵達終點，那就輸出路徑
        nextPath = GetNext(now);     #取得能走的位置                          
        for path in nextPath:
          sign = path.ToSequence();
          if (book.__contains__(sign) != True):#判斷當前節點狀態是否擴展過了
              #推入當前狀態，並標記該路徑已走過，因為每個狀態只需要擴展一次就夠了               
              queue.enqueue(path);
              book.setdefault(sign, []).append(True);
    #如果窮舉完都沒找到，代表無解
    print('無解')
    return None;
def PathTrace(now):
   path = list()
   while(now.father != None):
      path.append(now)
      now = now.father
   path.reverse()
   for i in path : print('執行步驟',i.step)
   return path 
def main():
  board = [0, 1, 2, 3, 4, 5, 6, 7, 8]
  random.shuffle(board)
  print("Board",board)
  globals = [1, 2, 3, 4, 5, 6, 7, 8, 0]
  print("Goal",globals)
  puzzle = Solve(board, globals)
if __name__ == "__main__":
    main()
```
